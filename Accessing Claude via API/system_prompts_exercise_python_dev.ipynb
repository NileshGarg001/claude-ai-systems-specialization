```python
# Showing Claude API does not store content / have inbuilt short term context 
from dotenv import load_dotenv
import os

# Load environment variables from .env file
load_dotenv()

# Create an API Client 
from anthropic import Anthropic
client = Anthropic()
model = "claude-sonnet-4-0"
```

```python
def add_user_message(messages,text):
    messages.append({"role": "user", "content": text})

def add_assistant_message(messages,text):
    messages.append({"role": "assistant", "content": text})

def chat(messages, system = "You are a helpful assistant."):
    message = client.messages.create(
        model=model,
        max_tokens=1000,
        messages = messages,
        system=system,
    )
    return message.content[0].text 
```

```python
# Without the system prompt 

messages = []
add_user_message(messages, "Give me the code for fibonacci series in python.")
response = chat(messages)
print(response)
```

```python
# With the system prompt. Try to generate the code only nothing else 
python_system_inst = """
You are an expert Python function developer. 
You speak and write in Python only. 
You goal is to generate just the python code with no explanations or comments. 
In order to explain logic you create the function documentation string"""
messages = []
add_user_message(messages, "Give me the code for fibonacci series in python.")
response = chat(messages, python_system_inst)
print(response)
```
